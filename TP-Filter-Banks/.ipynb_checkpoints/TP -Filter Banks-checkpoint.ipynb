# Functions

```python
def load_sound(file):
    return wave.open(file, 'rb')


def play_sound(file, chunk = 1024):
    """
    Script from PyAudio doc
    """
    wf = load_sound(file)
    p = pyaudio.PyAudio()
    stream = p.open(format=p.get_format_from_width(wf.getsampwidth()),
                    channels=wf.getnchannels(),
                    rate=wf.getframerate(),
                    output=True)
    data = wf.readframes(chunk)

    while data:
        stream.write(data)
        data = wf.readframes(chunk )

    stream.stop_stream()
    stream.close()
    p.terminate()
    
    
def plot_sound(data, times, name='default_name', save=False):
    plt.figure(figsize=(30, 4))
    plt.fill_between(times, data)
    plt.xlim(times[0], times[-1])
    plt.xlabel('time (s)')
    plt.ylabel('amplitude')
    if save:
        plt.savefig(name+'.png', dpi=100)
    plt.show()
```

# 1. Two-channel filter bank

```python
#1.1

# Defines the Remez filter params
N = 32                  # Size of the filter v
trans_width = 1/80      # Transition width
fc = 1/4            # Cutoff frequency

# Obtains the Remez filter
g = signal.remez(2*N-1, [0, fc, fc+trans_width, 1/2], [1, 0])

# Chebyshev’s alternance property 
v, G = signal.freqz(g)
v = v/(2*np.pi)
print("On the plot shown below we can clearly see that the Chebyshev’s alternance property holds, since the filter's oscilations have the same δ.\n")

plt.figure(figsize=(15, 4))
plt.plot(v, np.abs(G))
plt.title("Frequency response of g")
plt.xlim(0, 0.5)
plt.ylabel("Magnitude")
plt.xlabel("Frequency [(rad/sample)/2π]")
plt.grid()
plt.plot()

```

```python
# Check of the half band condition
print("\nOn the plot shown below we can appreciate that the half-band condition doesn't hold since one out of two coefficients of g(n) is not zero.\n")
plt.figure(figsize=(15, 4))
plt.title('Plot of g')
plt.xlim(0, len(g)-1)
plt.grid()
plt.plot(g, 'bo')
plt.xlabel('Index')
plt.ylabel('Amplitude')
plt.stem(g)
plt.show()
```

```python
#1.2
#Does Chebyshev's alternance property still rigorously hold?
#The half band condition is now achieved

ghb = np.zeros(len(g))
ghb[::2] = g[::2]
ghb[N-1] = g[N-1]

plt.figure(figsize=(15, 4))
plt.title('Plot of g half band')
plt.xlim(0, len(ghb)-1)
plt.grid()
plt.plot(ghb, 'bo')
plt.xlabel('Index')
plt.ylabel('Amplitude')
plt.stem(ghb)
plt.show()

v, Ghb = signal.freqz(ghb)
v = v/(2*np.pi)
print("On the plot shown below we can clearly see that the Chebyshev’s alternance property doesn't hold since the filter's oscilations don't have the same δ.\n")
plt.figure(figsize=(15, 4))
plt.plot(v, np.abs(Ghb))
plt.title("Frequency response of g half band")
plt.xlim(0, 0.5)
plt.ylabel("Magnitude")
plt.xlabel("Frequency [(rad/sample)/2π]")
plt.grid()
plt.plot()
```

```python
#1.3
fc = 0.48 

# Obtains the nearly all-pass type II FIR filter
vR = signal.remez(N, [0, fc, 1/2, 1/2], [1, 0])

# Obtains the LPF derived from the filter vR
c = 1                                                                    # Defines the constant c
delta = np.zeros(2*N-1)                      
delta[N-1] = 1              # Obtains the dirac δ(n - (N-1))
v = np.zeros(2*N-1)
v[::2] = vR                 #V(z^2) in time
g = c*(v+delta)                         # g(n)

v, G = signal.freqz(g)
v = v/(2*np.pi)

# Plots the obtained filter
plt.figure(figsize=(15, 4))
plt.title('Plot of g')
plt.xlim(0, len(g)-1)
plt.grid()
plt.plot(g, 'bo')
plt.xlabel('Index')
plt.ylabel('Amplitude')
plt.stem(g)
plt.show()

plt.figure(figsize=(15, 4))
plt.plot(v, np.abs(G))
plt.xlim(0, 0.5)
plt.title("Frequency response of g")
plt.ylabel("Magnitude")
plt.xlabel("Frequency [(rad/sample)/2π]")
plt.grid()
plt.plot()
```

```python
#1.4 
#PREGUNTAR COMO PLOTEAR LA PARTE NEGATIVA

v, G = signal.freqz(g)
v = v/(2*np.pi)

#Obtain GR from G: GR = G z^(N-1)
arg = np.arange(0, np.pi, np.pi/len(v))
delay = (np.cos(arg)+1j*np.sin(arg))**(N-1)
GR = G * delay

plt.figure(figsize=(15, 4))
plt.plot(v, np.abs(GR))
plt.xlim(0, 0.5)
plt.title("Frequency response of gR")
plt.ylabel("Magnitude")
plt.xlabel("Frequency [(rad/sample)/2π]")
plt.grid()
plt.plot()
```

It is not possible to factorize $G_R(\nu)$ in the form $G_R(\nu) = \tilde{H}_0(e^{j2\pi\nu})H_0(e^{j2\pi\nu})$ since $G_R$ is not always positive and the product does.

# 1.5) Raising of the half-band filter

```python
# Obtains the δ of the Chebyshev’s alternance property of the filter gr
start_stopband = np.where(v > 0.30)[0][0]
end_stopband = len(v)
epsilon = max(np.abs(GR[start_stopband:end_stopband]))+0.0001

# Shows the δgr obtained in the
#print(f"The obtained δgr was {δgr}, which can be clearly seen highlighted in the plot below.\n")
#gr.freqz_plot(db=False, absolute=False, yhighlights=[-δgr, 0, δgr])

# Obtains the filter g+
gplus = g.copy()
gplus[N-1] = g[N-1] + epsilon

# Plots the obtained filter
plt.figure(figsize=(15, 4))
plt.title('Plot of g+')
plt.xlim(0, len(g)-1)
plt.grid()
plt.plot(g, 'bo')
plt.xlabel('Index')
plt.ylabel('Amplitude')
plt.stem(g)
plt.show()
```

```python
print(gplus[N-1] - g[N-1])
```

```python
v, Gplus = signal.freqz(gplus)
v = v/(2*np.pi)

plt.figure(figsize=(15, 4))
plt.plot(v, np.abs(Gplus))
plt.xlim(0, 0.5)
plt.title("Frequency response of g+")
plt.ylabel("Magnitude")
plt.xlabel("Frequency [(rad/sample)/2π]")
plt.grid()
plt.plot()
```

Note that $g_+$ still satisfies the half-band condition for a slightly different value of $c$.

# 1.6

```python
#mostrar que la fase de h0 no es lineal
# Tolerance of the fixed point method
tolerance = 5E-4

# Initialization
h0 = np.zeros(N)
h0[0] = 1

column0 = np.zeros(N)
column0[N-1] = h0[0]
H = hankel(column0, h0)
gplusN = gplus[0:N]

# First error
error = np.linalg.norm(H@h0 - gplusN)

# Iterates until obtain an acceptable result
while error > tolerance:
    h0 = np.linalg.inv(H)@gplusN
    h0 = np.sqrt(np.abs(gplusN[N-1]))*h0/np.linalg.norm(h0) #h0 normalization
    column0[N-1] = h0[0]     #bc h0 is updated
    H = hankel(column0, h0)
    error = np.linalg.norm(H@h0 - gplusN)            

#h0 found!
v, H0 = signal.freqz(h0)
v = v/(2*np.pi)

plt.figure(figsize=(15, 4))
plt.plot(v, np.abs(H0))
plt.xlim(0, 0.5)
plt.title("Frequency response of h0")
plt.ylabel("Magnitude")
plt.xlabel("Frequency [(rad/sample)/2π]")
plt.grid()
plt.plot()
```

# 1.7 Implement the CQF filter bank

From $h_0$ we can then obtain: PREGUNTAR

$f_0(n) = h_0(N-1-n)$

$h_1(n) = (-1)^n h_0(N-1-n)$

$f_1(n) = (-1)^{n+1}h_0(n)$

```python
f0 = np.flip(h0)

signo = np.zeros(N)

for i in range(0,N):
    signo[i] = (-1)**i

h1 = signo*np.flip(h0)

f1 = -signo*h0
```

Read and plot the signal.

```python
data_path = os.getcwd()
filename = 'homme3.wav'
sound = os.path.join(data_path, filename) 

wavefile = load_sound(sound)
print(wavefile.getparams())
```

```python
Fs = int(wavefile.getframerate())
num_samples = int(wavefile.getnframes())
data = wavefile.readframes(num_samples)
data = struct.unpack('{n}h'.format(n=num_samples), data)
x = np.array(data)

timestep = 1/float(Fs)
times = np.arange(len(x))*timestep
plot_sound(x, times)
```

```python
play = False
if play :
    play_sound(sound) 
```

Filtering

```python
#channel 0 of CQF
s0t = signal.lfilter(h0,1, x)     #\tilde{s_0}
s0 = s0t[::2]                     #downsampling
s0h = np.zeros(2*len(s0))         #\hat{s_0}
s0h[::2] = s0                     #upsampling

x0 = signal.lfilter(f0,1, s0h)

#channel 1 of CQF
s1t = signal.lfilter(h1,1,x)     #\tilde{s_1}
s1 = s1t[::2]                    #downsampling
s1h = np.zeros(2*len(s1))        #\hat{s_1}
s1h[::2] = s1                    #upsampling

x1 = signal.lfilter(f1,1, s1h)
```

```python
#Reconstruction
xh = x0+x1

#amplification correction
xh = xh/max(xh)*max(x)
```

```python
print('Mean difference =', np.mean(xh-x))
print('Standard deviation difference', np.std(xh)-np.std(x))
```

We can observe that there is not much difference between the signal statistics.

```python
print('Original signal')
plot_sound(x, times)
print ('Reconstructed signal')
plot_sound(xh, times)
```

2.

We assume that $C(z)$ is of the form $1-\alpha z^{-1}$.
In order to keep the perfect reconstruction condition property at the output of the transmultiplexer, we just introduce a causal filter $D(z)$ at the output of $C(z)$ such that $C(z)D(z) = \beta z^{-n_0}$, where $n_0$ is an odd number.

Then, we get $D(z) = \frac{\beta z^{-n_0}}{1-\alpha z^{-1}}$ or what it is the same $d(n) = \beta \alpha ^{n-n_0} u(n-n_0)$, where $u(n)$ is the Heaviside step function.

```python
#Define D
alpha = 0.9
n0 = 1
beta = 1

d = np.zeros(N)
for n in range(n0,N):
    d[n] = beta*(alpha**(n-n0))

#Define C
c = np.zeros(N)
c[0] = 1
c[1] = -alpha
```

```python
#Encoder
s0t = np.zeros(2*len(s0))         #\tilde{s_0}
s0t[::2] = s0
s0tf = signal.lfilter(f0,1, s0t)

s1t = np.zeros(2*len(s1))         #\tilde{s_1}
s1t[::2] = s1
s1tf = signal.lfilter(f1,1, s1t)

stf = s0tf+s1tf
```

```python
#Channel
stc = signal.lfilter(c,1, stf)
x = signal.lfilter(d,1, stc)

#Decoder
s0h = signal.lfilter(h0,1, x)
s00 = s0h[::2]                     #downsampling

s1h = signal.lfilter(h1,1, x)
s11 = s1h[::2]                     #downsampling
```
